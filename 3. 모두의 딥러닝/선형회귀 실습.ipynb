```python
import numpy as np
x = [2,4,6,8]
y = [81,93,91,97]
```

```python
mx = np.mean(x)
my = np.mean(y)
```

```python
divisor = sum([(i-mx)**2 for i in x])
```

```python
def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i]-mx)*(y[i]-my)
    return d
diviend = top(x,mx,y,my)
```

```python
a = diviend/divisor
```

```python
b = my-(mx*a)
```

```python
print('분모 : ',divisor)
print('분자 : ',diviend)
print('기울기 a = ',a)
print('y절편 b = ',b)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [[2,81],[4,93],[6,91],[8,97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.show()
x_data = np.array(x)
y_data = np.array(y)
print(x_data)
print(y_data)
    
```

```python

a = 0
b = 0

lr = 0.05
epochs = 2001
for i in range(epochs):
    y_pred = a * x_data + b
    error = y_data-y_pred
    print('error = ', error)
    a_diff = -(1/len(x_data)) * sum(x_data*(error))
    b_diff = -(1/len(x_data)) * sum(y_data - y_pred)
    print('a_diff = ', a_diff)
    print('b_diff = ', b_diff)
    a = a-lr*a_diff
    b = b-lr*b_diff
    
    if i % 100 == 0 :
        print('epoch=%.f 기울기 =%.04f, 절편=%.04f' %(i,a,b))

y_pred = a*x_data+b
plt.scatter(x,y)
plt.plot([min(x_data),max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```

```python
## y_pred를 a로 편미분 = a_diff
## y_pred를 b로 편미분 = b_diff
## 함수의 n 제곱의 미분 {f(x)^n}' => n{f(x)}^n-1 * f'(x))
```
