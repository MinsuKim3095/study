```python
## y는 f*g = g(f(x))이다
## 이를 합성함수의 미분으로 보면,
## dy/dx = g'(f(x))f'(x)
##위 식에서 g'f(x)는 k의 변화량에 따른 y의 변화량, 즉 dy/dk 이며
## f'(x)는 x의 변화량에 따른 k의 변화량, dk/dx 이기때문에
## dy/dk * dk/dx를 통해, (chain rule)
## dy/dx를 구할 수 있게 된다
```
